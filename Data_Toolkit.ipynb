1. What is NumPy, and why is it widely used in Python?
 - NumPy is a Python library used for numerical computing. It provides fast multidimensional arrays, mathematical functions, and tools for linear algebra, making computations efficient and memory-optimized.

2. How does broadcasting work in NumPy?
 - Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding smaller arrays to match larger ones without copying data.

3. What is a Pandas DataFrame?
 - A DataFrame is a 2-dimensional, labeled data structure in Pandas that stores data in rows and columns, similar to a table or spreadsheet.

4. Explain the use of the groupby() method in Pandas
 - groupby() is used to split data into groups, apply functions (sum, mean, count), and combine results for data aggregation and analysis.

5. Why is Seaborn preferred for statistical visualizations?
 - Seaborn provides high-level statistical plots, attractive default styles, and built-in support for Pandas DataFrames.

6. Differences between NumPy arrays and Python lists
- NumPy Array	                                      Python List

7. What is a heatmap, and when should it be used?
 - A heatmap displays data using color intensity and is commonly used to visualize correlation matrices and patterns in large datasets.

8. What is a “vectorized operation” in NumPy?
 - Vectorized operations apply calculations to entire arrays at once, avoiding loops and improving performance.

9. How does Matplotlib differ from Plotly?
 - Matplotlib: Static plots, publication-ready
 - Plotly: Interactive, web-based visualizations

10. Significance of hierarchical indexing in Pandas
 - Hierarchical indexing (MultiIndex) allows multiple levels of indexing, enabling complex data representation and analysis.

11. Role of Seaborn’s pairplot() ?
 - pairplot() visualizes pairwise relationships between numerical variables and distributions on the diagonal.

12. Purpose of describe() in Pandas ?
 - describe() provides summary statistics like count, mean, std, min, max, and quartiles.

13. Why is handling missing data important in Pandas?
 - Missing data can distort results. Pandas provides methods like fillna() and dropna() to ensure accurate analysis

14. What are the benefits of using Plotly for data visualization?
 - Plotly allows creation of interactive and dynamic visualizations. It supports zooming, hovering, tooltips, and real-time updates, making it ideal for dashboards and web-based analytics.

15. How does NumPy handle multidimensional arrays?
 - NumPy uses the ndarray (N-dimensional array) object, which stores data in a contiguous memory block and allows efficient operations on 1D, 2D, and higher-dimensional arrays.

16. What is the role of Bokeh in data visualization?
 - Bokeh is used to create interactive, browser-based visualizations that support real-time data streaming and are suitable for web applications and dashboards.

19. How does Pandas simplify time series analysis?
 - Pandas provides date-time indexing, resampling, rolling windows, shifting, and easy time-based slicing for efficient time series data analysis.

20. What is the role of a pivot table in Pandas?
 - A pivot table summarizes large datasets by grouping, aggregating, and reshaping data, similar to Excel pivot tables.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
 - NumPy slicing returns a view of the data instead of a copy, making it faster and more memory-efficient than Python list slicing.

22. What are some common use cases for Seaborn?
 - Distribution plots (histogram, KDE)
 - Correlation heatmaps
 - Regression analysis
 - Categorical data visualization

```python
#1️ Create a 2D NumPy array and calculate the sum of each row
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])

row_sum = np.sum(arr,axis=1)
print(row_sum)
```

```python
#2 Pandas script to find the mean of a specific column
import pandas as pd

df = pd.DataFrame({'Age':[10,20,30,40]})
mean_age = df['Age'].mean()
print(mean_age)
```

```python
#3 Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]

plt.scatter(x,y)
plt.show()
```

```python
#4 Calculate correlation matrix using Seaborn and visualize with heatmap
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame ({
    'A':[1,2,3,4],
    'B':[4,3,2,1],
    'C':[2,3,4,5]})

corr = data.corr()
sns.heatmap(corr, annot= True)
plt.show()
```

```python
#5 Generate a bar plot using Plotly
import plotly.express as px
df = px.data.gapminder().query('year == 2026').head()

fig = px.bar(df, x='country', y='pop')
fig.show()
```

```python
#6 Create a DataFrame and add a new column based on an existing column
import pandas as pd

df = pd.DataFrame({'Salary': [20000, 30000, 40000, 50000]})

df['Bonus'] = df['Salary'] * 0.1

print(df)
```

```python
#7 Element-wise multiplication of two NumPy arrays
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

result = a * b
print(result)
```

```python
#8 Line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 15, 25, 35]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
#9 Generate a DataFrame and filter rows where a column value > threshold ?
import pandas as pd

df = pd.DataFrame({'Age':[10,20,30,40]})
threshold = 20
filtered_df = df[df['Age'] > threshold]
print(filtered_df)
```

```python
#10 Create a histogram using Seaborn?

import seaborn as sns
import matplotlib.pyplot as plt

data = [10,20,30,40,50]

sns.histplot(data, bins=5)
plt.show()
```
